## Implementacion y visualizacion de la transformada de Fourier Discreta (DFT) (solo porque se ve fachera)
En practicamente ningun caso se utilizaria la DFT existiendo la Fast Fourier Transform (FFT) que es mucho mas eficiente. Pero la DFT es muy linda de ver.

### Ecuaciones
Si tenemos un vector de muestras $f = \begin{bmatrix} f_1 & f_2 & \dots & f_n \end{bmatrix}^T$ la DFT nos permite obtener otro vector $\hat{f} = \begin{bmatrix} \hat{f}_1 & \hat{f}_2 & \dots & \hat{f}_n \end{bmatrix}^T$ donde $\hat{f_k}$ es la contribucion de la frecuencia $k$ en la señal original. La DFT y su inversa estan dadas por las siguientes ecuaciones:
$$\hat{f}_k = \sum_{j=0}^{n-1} f_j e^{-2 \pi i j k/n}$$

$$f_k = \frac{1}{n} \sum_{j=0}^{n-1} \hat{f}_j e^{2 \pi i j k/n}$$
Luego, como la DFT es un operador lineal, puede ser representada como un producto matricial:
$$\begin{bmatrix} \hat{f}_1 \\ \hat{f}_2 \\ \vdots \\ \hat{f}_{n} \end{bmatrix} = 
\begin{bmatrix} 1 & 1 & 1 & \dots & 1 \\ 
1 & \omega & \omega^2 & \dots & \omega^{n-1} \\ 
1 & \omega^2 & \omega^4 & \dots & \omega^{2(n-1)} \\ 
\vdots & \vdots & \vdots & \ddots & \vdots \\ 
1 & \omega^{n-1} & \omega^{2(n-1)} & \dots & \omega^{(n-1)(n-1)} 
\end{bmatrix} 
\begin{bmatrix} f_1 \\ f_2 \\ \vdots \\ f_n \end{bmatrix}$$
donde $\omega_n = e^{-2 \pi i / n}$ es la raiz $n$-esima de la unidad. 

```python
import numpy as np
import matplotlib.pyplot as plt 

# Construir la matriz DFT
n = 256
w = np.exp(-1j*2*np.pi/n)
J,K = np.meshgrid(np.arange(n),np.arange(n))
DFT = np.power(w, J*K)
ReDFT = np.real(DFT)
ImDFT = np.imag(DFT)
```

```python
# Parte real
plt.figure(figsize=(10,10))
plt.imshow(ReDFT, cmap='jet')
plt.colorbar()
plt.title(r'$\mathrm{Re}(DFT)$')
```

```python
# Parte imaginaria
plt.figure(figsize=(10,10))
plt.imshow(ImDFT, cmap='jet')
plt.colorbar()
plt.title(r'$\mathrm{Im}(DFT)$')
plt.show()
```
